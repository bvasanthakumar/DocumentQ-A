```python
from PyPDF2 import PdfReader
from typing_extensions import Concatenate

pdfreader = PdfReader("/content/Data-Modernization-Cloud-data-foundation.pdf")

raw_text = ""

#Read Text from PDF
for i, page in enumerate(pdfreader.pages):
  content = page.extract_text()
  if content:
    raw_text += content
```

```python
raw_text
```

```python
import re

clean_text = re.sub(r'[^\w\s]','',raw_text)

clean_text
```

```python
from langchain.text_splitter import CharacterTextSplitter

text_splitter = CharacterTextSplitter(
    separator = "\n",
    chunk_size = 800,
    chunk_overlap = 200,
    length_function = len
)

texts = text_splitter.split_text(clean_text)
```

```python
from sentence_transformers import SentenceTransformer
from langchain.vectorstores import FAISS
from langchain.embeddings import HuggingFaceEmbeddings
import faiss
import numpy as np

embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-V2")
chunk_text = [chunk for chunk in texts]
embeddings = (embedding_model.embed_documents(chunk_text))

document_search = FAISS.from_texts(chunk_text, embedding_model)
```

```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer
from langchain.chains import LLMChain
from langchain_core.prompts import PromptTemplate
from langchain_core.language_models import LLM

# Load TinyLlama Model and Tokenizer
model_name = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"

device = "cuda" if torch.cuda.is_available() else "cpu"  # Auto-detect GPU
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    device_map={"": device},  # Run on GPU if available
    torch_dtype=torch.float16 if device == "cuda" else torch.bfloat16  # Use FP16 for GPU
).to(device)  # Move model to GPU
```

```python
# Function to retrieve relevant chunks
def retrieve_relevant_chunk(query, k=3):
    results = document_search.similarity_search(query, k=k)
    return " ".join([doc.page_content for doc in results])

# Example usage
query = "What are the key challenges faced by the clients?"
relevant_context = retrieve_relevant_chunk(query, k=1)

print(relevant_context)
```

```python
# Define Custom LangChain Wrapper for TinyLlama
class TinyLlamaLLM(LLM):
    def _call(self, prompt: str, stop=None):
        inputs = tokenizer(prompt, return_tensors="pt").to(device)  # Move input to GPU
        with torch.no_grad():
            output = model.generate(**inputs, max_new_tokens=100)
        return tokenizer.decode(output[0], skip_special_tokens=True)

    @property
    def _llm_type(self):
        return "tinyllama"
```

```python
# Initialize Custom LLM
llm = TinyLlamaLLM(max_tokens=500)

# Example Inputs
context = relevant_context
question = "What are the key challenges faced by the clients?"

qa_prompt = PromptTemplate.from_template(
    f"""
Context:
{context}

Question:
{question}

Provide a structured, concise response with bullet points without repetition.
**Only answer the question directly without adding extra details or solutions.**
"""
)

qa_chain = LLMChain(llm=llm, prompt=qa_prompt)


formatted_prompt = f"""
Context:
{context}

Question:
{question}

Provide a structured, concise response with bullet points and asnw without repetition.
**Only answer the question directly without adding extra details or solutions.**

"""

answer = qa_chain.invoke({"context": formatted_prompt, "question": question})

print(answer)
```
